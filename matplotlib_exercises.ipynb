```python
import matplotlib.pyplot as plt
import math
```

```python
x1 = list(range(0, 100))
y1 = [(xa ** 2 - xa + 2) for xa in x1]

plt.plot(x1,y1)
plt.annotate('Origin', xy = (0,0), xytext = (-.3, 10))
plt.scatter(0,2)
```

```python
x2 = list(range(0,100))
y2 = [xb ** 0.5 for xb in x2]

plt.plot(x2,y2)
```

```python
x3 = list(range(0,100))
y3 = [xc ** 3 for xc in x3]

plt.plot(x3,y3)
```

```python
x4 = list(range(0,100))
y4 = [2 ** xd for xd in x4]

plt.plot(x4,y4)
```

```python
x5 = list(range(0,100))
y5 = [(1 / (xe + 1)) for xe in x5 if xe != -1]

plt.plot(x5,y5)
```

```python
fig = plt.figure(figsize = (10, 6))

subplot1 = fig.add_subplot(2,2,1)
subplot1.plot(x2,y2, c = 'red', label = '$y = \sqrt{x}$')

subplot2 = fig.add_subplot(2,2,2)
subplot2.plot(x3,y3, c = 'blue', label = '$y = x ^ 3$')

subplot3 = fig.add_subplot(2,2,3)
plt.plot(x4, y4, c = 'green', label = '$y = 2^x$')

subplot4 = fig.add_subplot(2,2,4)
plt.plot(x5,y5, c = 'yellow', label = '$y = 1 / (x+1)$')

fig.suptitle('matplotlib exercises')
fig.legend(loc='upper right', bbox_to_anchor = (1.1, .9))
plt.show()
```
