```python
import matplotlib.pyplot as plt
import math
```

```python
x = list(range(1, 5))
y1 = [1 for x1 in x]
y2 = [(math.log(x2)) for x2 in x]
y3 = [x3 for x3 in x]
y4 = [(x4 * (math.log(x4))) for x4 in x]
y5 = [(x5 ** 2) for x5 in x]
y6 = [(2 ** x6) for x6 in x]
y7 = [(math.factorial(x7)) for x7 in x]
y8 = [(x8 ** x8) for x8 in x]
ylist = [y1, y2, y3, y4, y5, y6, y7, y8]
```

```python
plt.plot(x,  y1, label = '$0(1)$')
plt.plot(x,  y2, label = '$O(log n)$')
plt.plot(x,  y3, label = '$O(n)$')
plt.plot(x,  y4, label = '$O(n log n)$')
plt.plot(x,  y5, label = '$O(n^2)$')
plt.plot(x,  y6, label = '$O(2^n)$')
plt.plot(x,  y7, label = '$O(n!)$')
plt.plot(x,  y8, label = '$O(n^n)$')
plt.legend(loc='upper right',  bbox_to_anchor=(1.3,1))
plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
```
